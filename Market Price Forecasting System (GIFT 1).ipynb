```python
# Task 1 — Create the Variables

Commodity = "Maize"
Market_Location = "Kano Central Market"
Yesterday_Price = 58000
Today_Price = 61500
Expected_Market_Price = 60000
Quantity_Available = 420
Additional_Supply_Received = 80
Minimum_Supply_Requirement = 450
Market_Open = True
```

```python
# Task 2 — Calculate the Price Difference
Price_Difference = Today_Price - Yesterday_Price

print("Price Difference:", "₦", Price_Difference)
```

```python
#Task 3 — Calculate the Percentage Price Change
Percentage_Price_Change = (Price_Difference / Yesterday_Price) * 100
print(f"Percentage Price Change: {Percentage_Price_Change:,.2f}%")
print("Percentage Price Change:", round(Percentage_Price_Change, 2),"%")
```

```python
#Task 4 — Update Available Supply
Quantity_Available += Additional_Supply_Received
print("Quantity Available:", Quantity_Available, "bags")
```

```python
#Task 5 — Check if the Expected Market Price Has Been Reached

Expected_Price_Reached = Today_Price >= Expected_Market_Price
print("Expected Price Reached:", Expected_Price_Reached)
```

```python
#Task 6 — Check if Supply Is Sufficient
print("Supply Sufficient:", Quantity_Available >= Minimum_Supply_Requirement)
Sufficient_Supply = Quantity_Available >= Minimum_Supply_Requirement
```

```python
#Task 7 — Determine if Market Conditions Are Favorable
print("Market Favourable:", Market_Condition)
Market_Condition = Market_Open and Expected_Price_Reached
```

```python
#Task 8 — Determine if Market Monitoring Is Required
Market_Monitoring_Required = (not Market_Condition) or (not Sufficient_Supply)
print("Market Monitoring Required:", Market_Monitoring_Required)
```

```python
#Task 9 — Generate the Market Price Forecast Report

print(f"""
==============================================================
          AGRICULTURAL MARKET PRICE ANALYSIS REPORT
==============================================================

Commodity: {Commodity}

Market Location: {Market_Location}

Yesterday's Price:  ₦{Yesterday_Price}

Today's Price: ₦{Today_Price}

Price Difference: ₦{Price_Difference}

Percentage Price Change: {Percentage_Price_Change:,.2f}%

Updated Quantity Available: {Quantity_Available} bags

Minimum Supply Requirement: {Minimum_Supply_Requirement} bags

Expected Market Price: ₦{Expected_Market_Price}

Expected Price Reached: {Expected_Price_Reached}

Supply Sufficient: {Sufficient_Supply}

Market Open: {Market_Open}

Market Conditions Favorable: {Market_Condition}

Additional Market Monitoring Required: {Market_Monitoring_Required}

==============================================================""")
```
